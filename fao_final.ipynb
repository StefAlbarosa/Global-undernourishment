## food_supply_kcal and food_supply_kgprotein

I just worked my way backwards and from kcal/cap/day we can just get the supply of kcal of an item per year.

## food_supply_kg

```python
data.columns
```

The value figures of 'food' are in units of 1000 metric tons. To get kgs, I times this by 1000 X 1000.

```python
#but the label sais kg/capita/yr?
# sum the column just to look at it.
data['food_supply_kg']= data['food']* 1000000
data['food_supply_kg'].sum()
```

## ratio_kcalkg and protein_percentage
Here I just divided the caloric and protein values with the kg values of every item.

```python
# I had to add this part replace(np.inf, np.nan)  becaue the ratio_kcalkg was giving inf in several places 
data['ratio_kcalkg']=(data['food_supply_kcal']/data['food_supply_kg']).replace(np.inf, np.nan)
data['protein_percentage']=(data['food_supply_kgprotein']/data['food_supply_kg']*100).replace(np.inf, np.nan)  

display(data[['country','item','food_supply_kgprotein', 'food_supply_kg', 'ratio_kcalkg', 'protein_percentage']].head())

```


## dom_sup_kcal and dom_sup_kgprot

I multiplied the domestic supply values by 1000000 and by the ratio to get the domestic supply in terms of kcal and protein values.

```python
data['dom_sup_kcal'] = data['domestic_supply_quantity'] *1000000 * data['ratio_kcalkg']
data['dom_sup_kgprot'] = data['domestic_supply_quantity'] * 1000000 * data ['protein_percentage']

data.head()
```

How many humans on earth could be fed if all the plant-based


I summed from data all of food and feed and then multiplied by 10000000 and then multiplied by the protein_percentage and by .01 to get the actual number not the percent.
